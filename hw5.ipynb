```python
# Подключение библиотек
import sqlite3
import pandas as pd
```

```python
# Подключение к БД
con = sqlite3.connect('book_simple.db')
cur = con.cursor() 
```

```python
# Ваши запросы здесь
q_tables = """
SELECT 
    name
  FROM 
    sqlite_master;
"""
cur.execute(q_tables).fetchall()
```

Выведите авторов, имеющихся в базе

```python
q_authors = """
SELECT 
    name
  FROM 
    author;
"""
pd.read_sql(q_authors, con) 
```

Найдите всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия

```python
q_authors_CA = """
SELECT 
    name
  FROM 
    author
  WHERE 
    name 
   LIKE 
    ('С. А.%');
"""

pd.read_sql(q_authors_CA, con) 
```

Найдите все издательства, название которых состоит из двух и более слов.

```python
q_publisher_2w = """
SELECT 
    name
  FROM 
    publisher
  WHERE 
    name 
   LIKE 
    '% %';
"""

pd.read_sql(q_publisher_2w, con) 
```

Выведите список книг, выпущенных в издательстве "Мир".

```python
q_publisher_mir = """
SELECT 
    publisher.name,
    book.title
  FROM 
    publisher,
    book 
  WHERE 
      book.publisher_id = publisher.id 
    AND 
      publisher.name = 'Мир';
"""

pd.read_sql(q_publisher_mir, con) 
```

Выведите количество книг, принадлежащих каждой из категорий. Результат отсортировать по возрастанию.

```python
q_categ_dist = """
SELECT DISTINCT 
      book.category_id
  FROM 
      book;
"""

pd.read_sql(q_categ_dist, con) 
```

```python
q_categ_count = """
SELECT 
       book.category_id AS category,
       COUNT(book.title) AS books
  FROM
       book
  GROUP BY
      book.category_id 
  ORDER BY books ASC;
"""

pd.read_sql(q_categ_count, con) 
```

Выведите количество книг, выпущенных в каждом издательстве. В итоговом списке оставьте только те издательства, которые выпустили 3 и более книг. Результат отсортировать по убыванию.

```python
q_categ_count2 = """

SELECT *
FROM (
      SELECT 
            publisher.name AS publisher,
            COUNT(book.title) AS books
        FROM
            publisher,
            book
        WHERE 
            book.publisher_id = publisher.id
        GROUP BY
            publisher 
        ORDER BY books DESC
      )
WHERE 
    books >= 3;
"""

pd.read_sql(q_categ_count2 , con) 
```

Выведите список книг, автором которых являлся <<С. А. Айвазян>>.

```python
q_vasyan = """
SELECT 
    name,
    title
  FROM
    authorbook
  INNER JOIN 
    book
  ON authorbook.book_id = book.id 
  INNER JOIN 
    author
  ON authorbook.author_id = author.id
  WHERE 
    name = 'С. А. Айвазян';
"""

pd.read_sql(q_vasyan, con) 
```

В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>. Результат отсортировать по убыванию.

```python
q_vasyan_caunter = """
SELECT
  author.name,
  publisher.name AS publisher,
  COUNT(book.title) AS count
FROM
  author
INNER JOIN
  authorbook
ON author.id = author_id
INNER JOIN
  book
ON book_id = book.id
INNER JOIN
  publisher
ON publisher_id = publisher.id
WHERE
  author.name = 'С. А. Айвазян'
GROUP BY
  publisher
ORDER BY count DESC
"""

pd.read_sql(q_vasyan_caunter, con) 
```
