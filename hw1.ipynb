# Практическое задание 1 (обязательное)

## Исходные данные

В словаре `dict_users_network` заданы взаимосвязи (`дружба`) пользователей социальной сети, где `ключ` -- имя пользователя; `значение` -- список друзей пользователя. Для простоты будем считать, что оношение `дружба` не является биективным, т.е. если у `A` другом является `B`, то не обязательно, что у `B` в друзьях `A`. 

## Задание (обязательное) 
- Создайте функцию, определяющую количество друзей у выбранного (каждого) пользователя.
- Создайте функцию, выводящую на экран пользователей, отсортированных по количеству друзей.
- Опубликуйте ссылку на файл расширения `ipynb`, прикрепив его в этом задании.

## Задание (факультативное)
- Создайте функцию, определяющую наличие однофамильцев.
- Создайте функцию, определяющую наличие пользователей с одинаковыми инициалами.
- Создайте функцию, возвращающую список уникальных пользователей сети в формате `Фамилия И.О.`.
- Создайте функцию, возвращающую список из уникальных фамилий  пользователей сети.

**Добавлен новый пользователь в качестве друга - *С.В. Шошкин* для проверки гибкости кода!**

```python
dict_users_network = ({
    'С.В. Сидоров' : ['П.В. Сидоров', 'И.И. Иванов', 'А.С. Никифоров', 
                      'Ф.С. Афеногонов', 'А.К. Парфенов'],
    'П.В. Сидоров' : ['С.В. Сидоров', 'А.С. Шебалова', 'М.Л. Кривов'],
    'И.И. Иванов' : ['Л.Т. Тимофеев', 'С.В. Сидоров', 'П.В. Сидоров', 
                     'М.Л. Кривов'],
    'А.С. Никифоров' : ['С.В. Сидоров', 'П.В. Сидоров'],                 
    'Ф.С. Афеногонов' : ['С.В. Сидоров', 'П.В. Сидоров', 'А.К. Парфенов'],
    'А.С. Шебалова' : ['Л.Т. Тимофеев'],
    'Л.Т. Тимофеев' : ['К.К. Петров', 'А.С. Крылов', 'С.В. Шошкин']
                     })
```

```python
# Функция, определяющая количество друзей у выбранного пользователя принимающая в качестве аргумента словарь с которым необходимо работать.
def friends_counter(d):
    while 1 == 1:
      print("Введите имя пользователя (или слово 'All' для отобоброжения всех пользователей):");
      inpt = input();

      if inpt in d:
        print(f"У пользователя с именем {inpt} в друзьях - {len(d[inpt])} человек(а), их имена - {d[inpt]} ");
        break;
      elif inpt == "All":
        for u in d:
          print(f"У пользователя с именем {u} в друзьях - {len(d[u])} человек(а), их имена - {d[u]} ");
        break;
      else: print(f"Пользователь с именем {inpt} не найден");

friends_counter(dict_users_network);
```

```python
# Функция, выполняющая сортировку пользователей по количеству друзей принимающая в качестве аргумента словарь с которым необходимо работать.
def sort_by_friends(d):
  print("Список пользователей отсортированных по количеству друзей:\n")
  for key, value in sorted(d.items(), key=lambda x:len(x[1])):
    print(f"{key}, {value}");

sort_by_friends(dict_users_network);
```

```python
# Функция, выполняющая поиск однофамильцев по всей социальной сети принимающая в качестве аргумента словарь с которым необходимо работать.

def surname_equal_search(d):
  list_unique_users = [];
  list_unique_users_clear = [];
  counter = {};

  for i, j in d.items():
    if i not in list_unique_users:
      list_unique_users.append(i);
    for k in j:
      if k not in list_unique_users:
        list_unique_users.append(k);

  for i in list_unique_users:
    list_unique_users_clear.append(i.split(" ")[-1]);

  for i in list_unique_users_clear:
    counter[i] = counter.get(i, 0) + 1

  doubles = {f: c for f, c in counter.items() if c > 1} 
  print(f"Количество однофамильцев в социальной сети:\n  {doubles}");

surname_equal_search(dict_users_network);
```

```python
# Функция, выполняющая поиск людей с одинаковыми инициалами по всей социальной сети принимающая в качестве аргумента словарь с которым необходимо работать.

def initial_equal_search(d):
  list_unique_users = [];
  split_unique_users = [];

  for i, j in d.items():
    if i not in list_unique_users:
      list_unique_users.append(i);
    for k in j:
      if k not in list_unique_users:
        list_unique_users.append(k);

  for i in list_unique_users:
    split_unique_users.append(i.split(" "));

  counter = 0;
  for i in range(0,len(split_unique_users)):
    for j in range(i,len(split_unique_users)):
      if split_unique_users[i][0] == split_unique_users[j][0] and split_unique_users[i][1] != split_unique_users[j][1]:
        counter += 1;
        print(f"{split_unique_users[i]} и {split_unique_users[j]} имеют одинаковые инициалы.");
  print(f"\nВсего: {counter} человек(а) с одинаковыми инициалами.");

initial_equal_search(dict_users_network);
```

```python
# Функция, возвращающая список уникальных пользователей сети в формате Фамилия И.О. принимающая в качестве аргумента словарь с которым необходимо работать.

def get_unique_list_users(d):
  list_unique_users = [];
  list_unique_users_reverse = [];

  for i, j in d.items():
    if i not in list_unique_users:
      list_unique_users.append(i);
    for k in j:
      if k not in list_unique_users:
        list_unique_users.append(k);

  for i in list_unique_users:
    list_unique_users_reverse.append(i.split(" ")[-1] + " " + i.split(" ")[0]);

  return list_unique_users_reverse;

print(get_unique_list_users(dict_users_network));
```

```python
# Функция, возвращающая список из уникальных фамилий пользователей сети принимающая в качестве аргумента словарь с которым необходимо работать.

def get_unique_list_users_surname(d):
  list_unique_users = [];
  unique_list_users_surname = [];

  for i, j in d.items():
    if i not in list_unique_users:
      list_unique_users.append(i);
    for k in j:
      if k not in list_unique_users:
        list_unique_users.append(k);

  for i in list_unique_users:
    unique_list_users_surname.append(i.split(" ")[-1]);

  return unique_list_users_surname;

print(get_unique_list_users_surname(dict_users_network));
```
