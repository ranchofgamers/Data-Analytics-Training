```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

link = 'https://devicedb.xyz/dev/category/24-msi-graphics-card/'
s = BeautifulSoup(requests.get(link).text, 'lxml')
max_page = int(s.find('li', class_ = 'ipsPagination_pageJump').\
                    find('li', class_ = 'ipsFieldRow').\
                    find_next().get("max"))
```

```python
schems = []

for i in range (1, max_page + 1):
  link_end = link + str(i)
  soup = BeautifulSoup(requests.get(link_end).text, 'lxml')
  items = soup.find_all('li', class_ = 'ipsDataItem')
  
  for item in items:
    schems.append(
        {
         'Автор' : item.find('p', class_ = 'ipsType_reset ipsType_light ipsType_blendLinks').get_text(strip=True)[2:],
         'Название': item.find('h4', class_ = 'ipsDataItem_title ipsType_sectionHead ipsContained_container').get_text(strip=True),
         'Описание' : item.find('div', class_ = 'ipsType_medium ipsType_richText ipsType_break ipsSpacer_top').get_text(strip=True),
        })

df = pd.DataFrame(schems)
df
```
